## üîç Key insights from the summary statistics:

- ‚úÖ Some columns (e.g., stargazers, forks) show high variance, with stargazers having a max value of 6351 while the median is only 45. This suggests a skewed distribution with a few repositories having extreme popularity.

- ‚úÖ Some features (e.g., desc_trigram_most_common_*) are sparse, with most values being 0.

- ‚úÖ Certain columns (e.g., releases) contain zero values, which may or may not be meaningful (e.g., repositories with no releases).

- ‚úÖ Features like branches, watchers, and forks have relatively low dispersion, making them potentially stable predictors.

```python
# Identify outliers using the Interquartile Range (IQR) method
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Count how many values are considered outliers per column
outlier_counts = ((df < lower_bound) | (df > upper_bound)).sum()

# Display only columns where outliers exist
outlier_counts[outlier_counts > 0]
```

# Why Log Transformation?

- ‚úÖ Reduces the impact of extreme values without losing information.

- ‚úÖ Improves model performance by making distributions more symmetric.

- ‚úÖ Preserves the ranking of values (i.e., a high value remains high but in a compressed range).

```python
import numpy as np

# Define the selected features for log transformation
log_features = ["forks", "watchers", "stargazers", "contributors", "branches"]

# Apply log1p transformation to selected columns
df[log_features] = df[log_features].apply(lambda x: np.log1p(x))

# Verify the transformation by checking the new distribution
df[log_features].describe().transpose()
```
